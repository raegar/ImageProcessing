```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image from file
image = cv2.imread('example_image.png')

# Convert the image from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
# Convert the image to grayscale for edge detection
gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Display the image
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title('Greyscale Image')
plt.show()

```

```python
# Apply Canny Edge Detection to the Greyscale image
edges = cv2.Canny(gray_image, threshold1=90, threshold2=110)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.title('Canny Edge Detection')
plt.show()
```

```python
# Apply Median Blurring
median_blurred_image = cv2.medianBlur(gray_image, 5)

# Display the median-blurred image
plt.imshow(median_blurred_image, cmap='gray')
plt.axis('off')
plt.title('Median Blurred Image')
plt.show()



```

```python
# Apply Canny Edge Detection to blurred image
edges = cv2.Canny(median_blurred_image, threshold1=100, threshold2=200)

# Display the edges
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.title('Canny Edge Detection')
plt.show()
```
