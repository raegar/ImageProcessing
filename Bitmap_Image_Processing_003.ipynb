```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image from the local file
image_path = 'Incredible.bmp'
image = Image.open(image_path)

# Convert the image to 8-bit grayscale
gray_image_8bit = image.convert('L')  # 'L' mode for 8-bit grayscale

# Save the grayscale image as an 8-bit BMP
gray_image_8bit.save("incredible_8bit.bmp", format="BMP")

# Display the original and grayscale images side by side
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.axis('off')
plt.title("Original Image: incredible.bmp")

# Grayscale image
plt.subplot(1, 2, 2)
plt.imshow(gray_image_8bit, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image (8-bit)")

plt.tight_layout()
plt.show()

print("Grayscale image saved as an 8-bit BMP: incredible_8bit.bmp")
```
