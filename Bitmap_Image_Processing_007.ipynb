```python
from PIL import Image
import matplotlib.pyplot as plt

def resize_image(input_path, output_path, new_size=(32, 32)):
    image = Image.open(input_path)

    resized_image = image.resize(new_size, Image.NEAREST)
    resized_image.save(output_path, format='BMP')
    print(f"Image successfully resized to {new_size} and saved as {output_path}")

#Load the image from the local file
image_path = 'Incredible.bmp'
image = Image.open(image_path)

input_image_path = 'Incredible_bw.bmp'
output_image_path = 'Incredible_32x32.bmp'

#Convert the image to black and white with dithering

bw_image = image.convert('1', dither=Image.FLOYDSTEINBERG)

bw_image.save('Incredible_bw.bmp', format='BMP')

# Convert the image to 8-colours
image_8_colour = image.convert('P', palette=Image.Palette.ADAPTIVE, colors=8)

# Save the 8 colour image as a PNG
image_8_colour.save("incredible_8.png", format="PNG")

resize_image(input_image_path, output_image_path)

plt.figure(figsize=(10, 5))

#Original image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.axis("off")
plt.title("Original image: Incredible.bmp")

#Greyscale image
plt.subplot(1, 2, 2)
plt.imshow(bw_image, cmap='gray')
plt.axis("off")
plt.title("Black and White image")

plt.tight_layout()
plt.show()

```
