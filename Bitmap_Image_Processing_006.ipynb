```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image from the local file
image_path = 'Incredible.bmp'
image = Image.open(image_path)

# Convert the image to 8-bit grayscale
image_8_colour = image.convert('P', palette=Image.Palette.ADAPTIVE, colors=8)

# Save the grayscale image as an 8-bit BMP
image_8_colour.save("incredible_8.png", format="PNG")

# Display the original and grayscale images side by side
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.axis('off')
plt.title("Original Image: incredible.bmp")

# Grayscale image
plt.subplot(1, 2, 2)
plt.imshow(image_8_colour)
plt.axis('off')
plt.title("8 Colour Web Safe Image")

plt.tight_layout()
plt.show()

print("8 Colour Web Safe Image saved as a PNG: incredible_8.png")
```
